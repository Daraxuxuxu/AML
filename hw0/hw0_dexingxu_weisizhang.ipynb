```python
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline

result = []  # N * p array, N is the number of samples and p is the number of attributes per sample
colors = []  # N dimensional vector for labels

for line in open("iris.csv"):
    line = line.strip().split(",")
    result.append(line[:-1])
    if(line[4] == 'Iris-setosa'):
        colors.append('r')
    if(line[4] == "Iris-versicolor"):
        colors.append('g')
    if(line[4] == "Iris-virginica"):
        colors.append('b')

plt.figure(figsize=(16, 16))
plt.suptitle('Iris Data(red=setosa,green=versicolor,blue=virginica)',fontsize=24)
name = ['Sepal.Length','Sepal.Width','Petal.Length', 'Petal.Width']

index = 1
for i in range(4):
    for j in range(4):
        plt.subplot(4,4,index)
        index += 1
        
        if (i != j):
            x = np.array(result)[:,i]
            y = np.array(result)[:,j]
            plt.title(name[i]+' x '+ name[j])
            plt.scatter(x, y, c=colors)
        else:
            plt.annotate(name[i], (0.5, 0.5), xycoords='axes fraction', ha='center', va='center')
plt.savefig("plot.png")
plt.show()
```
